Импорт библиотек

```python
import requests
from bs4 import BeautifulSoup
from time import sleep
import pandas as pd
```

```python
data = []


# цикл поиска по всем страницам. В данном случае взял только первые три. С каждым увеличением 
# обрабатываемых страниц, увеличивается вероятность ошибки связанной с запроса капчи
# 

for p in range(1, 4):
    
    # выводим шаги запросов
    print(p)
    
    # Запрос к сайту.
    
    url = f"https://www.litres.ru/luchshie-knigi/page-{p}/"
    reguest_site = requests.get(url)
    # избегаем перегрузки запросами, делаем паузы 
    sleep(3)
    # Обработка данных с помощью библиотеки BeautifulSoup
    soup = BeautifulSoup(reguest_site.text, 'lxml')
    
    bestsellers = soup.findAll('div', class_ = 'art-standart art-item mgrid cover-309')

    for bestseller in bestsellers:
        # Находим автора книги. Так есть книги без указания автора, 
        # чтобы избежать ошибок используем исключение (конструкцию try except) 
        try:
            author_book = bestseller.find('div', class_ = 'art__author').find('a').get('title')
        except:
            author_book = '-'
        # Находим наименование книги    
        name_book = bestseller.find('div', class_ = 'art__name').find('a', class_ = 'art__name__href').get('title')
        
        # Находим ссылку на книгу
        link = ("https://www.litres.ru" + bestseller.find('div', class_ = 'art__name').
                find('a', class_ = 'art__name__href').get('href'))
        
        # Находим оценку книги читателями
        
        rating_book = bestseller.find('div', class_ = 'inline-elem bottomline-rating').text
         
      
        # Переводи рейтинг из str в float
        rating_book = float(rating_book.replace(",", "."))
        
        # Заполняем список
        data.append([author_book, name_book, link, rating_book])
```

Проверяем список

```python
data
```

Создаем заголовки столбцов для dataframe

```python
header = ['Автор', 'Наименование', 'Ссылка', 'Оценка читателей']
```

Создаем dataframe и проверяем его

```python
df = pd.DataFrame(data, columns=header)
df
```

Записываем в файл csv

```python
df.to_csv('litres_bestsellers.csv')
```
